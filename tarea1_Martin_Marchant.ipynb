## `Pregunta 2)` **Cálculos y gráficos en el dataset general:**

### `A)`  Calcula el promedio, la desviación estándar, el valor mínimo, primer cuartil, mediana, tercer cuartil y máximo para las variables EDAD_ALU, PROM_GRAL y ASISTENCIA

#### `Respuesta:` A continuación se mostrarán los calculos solicitados para cada variable:

```python
variables = ['EDAD_ALU','PROM_GRAL','ASISTENCIA']
for i in variables:
    print(f'El promedio de {i} es {df_rendimiento[i].mean():.2f}')
    print(f'La desviación estandar de {i} es {df_rendimiento[i].std():.2f}')
    print(f'El valor mínimo de {i} es {df_rendimiento[i].min():.2f}')
    print(f'El primer cuartil de {i} es {df_rendimiento[i].quantile(0.25):.2f}')
    print(f'La mediana de {i} es {df_rendimiento[i].median():.2f}')
    print(f'El tercer cuartil de {i} es {df_rendimiento[i].quantile(0.75):.2f}')
    print(f'El valor máximo de {i} es {df_rendimiento[i].max():.2f}')
    print(f'')
```

### `B)`  La variable NOM_COM_ALU indica la comuna de residencia del alumno. Haz una tabla con la cantidad de estudiantes para cada comuna de residencia, ordenada desde la comuna con mayor cantidad de alumnos hasta la menor

#### `Respuesta:` A continuación se mostrarán la tabla con las cantidades de estudiantes por cada comuna de residencia:

```python
residencia_alumno = df_rendimiento.groupby('NOM_COM_ALU')['MRUN'].agg('count').reset_index().sort_values('MRUN',ascending=False)
residencia_alumno = residencia_alumno.rename(columns = {'MRUN':'Cantidad'})
residencia_alumno
```

### `C)`  La variable NOM_COM_RBD indica la comuna del establecimiento educacional del estudiante. Haz una tabla con la cantidad de estudiantes para cada comuna de residencia, ordenada desde la comuna con mayor cantidad de alumnos hasta la menor. Indica la comuna con mayor porcentaje de estudiantes residentes que estudian en un establecimiento de la misma comuna.

#### `Respuesta:` A continuación se mostrarán la tabla con las cantidades de estudiantes por cada comuna del establecimiento educacional:

```python
residencia_eduacion = df_rendimiento.groupby('NOM_COM_RBD')['MRUN'].agg('count').reset_index().sort_values('MRUN',ascending=False)
residencia_eduacion = residencia_eduacion.rename(columns = {'MRUN':'Cantidad_e','NOM_COM_RBD':'Comuna'})
residencia_eduacion
```

#### `Respuesta:` A continuación se mostrarán la tabla con el porcentaje de los estudiantes que viven y estudian en la misma comuna del total de los que estudian ahí.
#### La comuna con mayor porcentaje de estudiantes residentes que estudian en un establecimiento de la misma comuna es `Curanilahue`

```python
estudian_donde_viven = df_rendimiento[df_rendimiento.NOM_COM_RBD == df_rendimiento.NOM_COM_ALU]
estudian_donde_viven = estudian_donde_viven.groupby('NOM_COM_RBD')['MRUN'].agg('count').reset_index().sort_values('MRUN',ascending=False)
estudian_donde_viven = estudian_donde_viven.rename(columns={'MRUN':'Cantidad_v_e','NOM_COM_RBD':'Comuna'})
estudian_donde_viven = estudian_donde_viven.merge(residencia_eduacion, on = 'Comuna', how = 'left')
estudian_donde_viven['Porcentaje_estudiantes_viven_estudian'] = (estudian_donde_viven['Cantidad_v_e']/estudian_donde_viven['Cantidad_e'])*100
estudian_donde_viven.sort_values('Porcentaje_estudiantes_viven_estudian', ascending = False)
```
